```python
from duckduckgo_search import AsyncDDGS
from slugify import slugify
import aiohttp
import asyncio
import aiofiles
import os
from pathlib import Path
import logging
from typing import List, Dict
import time

class AsyncImageDownloader:
    def __init__(self, output_dir: str = 'logo_tenk', max_concurrent: int = 10):
        self.output_dir = Path(output_dir)
        self.output_dir.mkdir(exist_ok=True)
        self.max_concurrent = max_concurrent
        self.failed_downloads: List[Dict] = []
        
        # Setup logging
        logging.basicConfig(level=logging.INFO)
        self.logger = logging.getLogger(__name__)
        
    async def search_image(self, company: str) -> str:
        """Search for company logo using DuckDuckGo."""
        search_term = f'{company} company latest logo'
        try:
            async with AsyncDDGS() as ddgs:
                results = await ddgs.aimages(
                    search_term,
                    region='wt-wt',
                    safesearch='off',
                    max_results=1
                )
                return results[0]['image'] if results else None
        except Exception as e:
            self.logger.error(f"Search failed for {company}: {str(e)}")
            return None

    async def download_image(self, session: aiohttp.ClientSession, url: str, company: str) -> bool:
        """Download image from URL."""
        try:
            async with session.get(url, timeout=aiohttp.ClientTimeout(total=10)) as response:
                if response.status == 200:
                    content = await response.read()
                    output_path = self.output_dir / f"{company}.jpg"
                    async with aiofiles.open(output_path, 'wb') as f:
                        await f.write(content)
                    return True
                return False
        except Exception as e:
            self.logger.error(f"Download failed for {company}: {str(e)}")
            return False

    async def process_company(self, session: aiohttp.ClientSession, company: str):
        """Process a single company - search and download its logo."""
        try:
            image_url = await self.search_image(company)
            if image_url:
                success = await self.download_image(session, image_url, company)
                if not success:
                    self.failed_downloads.append({
                        'company': company,
                        'reason': 'Download failed'
                    })
            else:
                self.failed_downloads.append({
                    'company': company,
                    'reason': 'No image found'
                })
        except Exception as e:
            self.failed_downloads.append({
                'company': company,
                'reason': str(e)
            })

    async def download_all(self, companies: List[str]):
        """Download logos for all companies with rate limiting."""
        async with aiohttp.ClientSession() as session:
            tasks = []
            for company in companies:
                if len(tasks) >= self.max_concurrent:
                    # Wait for some tasks to complete before adding more
                    await asyncio.gather(*tasks)
                    tasks = []
                
                tasks.append(asyncio.create_task(
                    self.process_company(session, company)
                ))
                
                # Small delay to prevent overwhelming the API
                await asyncio.sleep(0.2)
            
            # Wait for remaining tasks
            if tasks:
                await asyncio.gather(*tasks)

# Usage example
async def main(companies):
    start_time = time.time()
    
    downloader = AsyncImageDownloader(
        output_dir='logo_all',
        max_concurrent=10
    )
    
    await downloader.download_all(companies)
    
    # Report results
    print(f"\nDownload completed in {time.time() - start_time:.2f} seconds")
    print(f"Failed downloads: {len(downloader.failed_downloads)}")
    for failure in downloader.failed_downloads:
        print(f"- {failure['company']}: {failure['reason']}")

companies = final_df['name'].tolist()
await main(companies)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
import shutil

# Define directories for train grouped by industry
train_dir = 'train'
path='logo_all'
errors = []
c = 0
# Function to split by industry and move files
for industry, group in final_df.groupby('industry'):
    train_industry_dir = os.path.join(path, industry)
    os.makedirs(train_industry_dir, exist_ok=True)
    
    # Move files to the appropriate directory based on the split
    for _, row in group.iterrows():
        image_name = f"{str(row['name'])}.jpg"  # Assuming images are .jpg
        src_path = os.path.join(path, image_name)  # Original images folder
        dst_path = os.path.join(train_industry_dir, image_name)
        if os.path.exists(src_path):
            try:
                shutil.copy(src_path, dst_path)
                c += 1
            except:
                errors.append(src_path)
print(len(errors))
```
